```python
import matplotlib.pyplot as plt
from skimage import io
```

```python
origImg = io.imread('./images/eagle.jpg')
absImg = io.imread('./images/eagle_abs.jpg')
uniqImg = io.imread('./images/eagle_uniq.jpg')
distImg = io.imread('./images/eagle_dist.jpg')
salImg = io.imread('./images/eagle_sal.jpg')

fig1,ax1= plt.subplots(1,5,figsize=(10,2))


#fig1.suptitle("Pipeline
ax1[0].set_title('Original')
ax1[0].imshow(origImg)

ax1[1].set_title('Abstracted')
ax1[1].imshow(absImg)

ax1[2].set_title('Uniqueness')
ax1[2].imshow(uniqImg)

ax1[3].set_title('Distribution')
ax1[3].imshow(distImg)

ax1[4].set_title('Saliency')
ax1[4].imshow(salImg)

# # ax1.set_ylabel('V')
# # ax1.set_xlabel('t')

# # Turn off tick labels
# ax1[3].set_yticklabels([])
# ax1[3].set_xticklabels([])


fig1.tight_layout()
#fig1.subplots_adjust(top=0.99)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)
plt.show()
fig1.savefig("./output_data/pipeline2.jpg")
```
